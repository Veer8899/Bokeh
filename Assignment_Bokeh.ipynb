## **Q1. How can you create a Bokeh plot using Python code?**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, show, output_file
```

```python
p =figure(title = 'data')
```

```python
x = [1, 2, 3, 4, 4.5]
y = [6, 7, 2, 4, 5]
p.line(x,y)
show(p)
```

## **Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.**

### In Bokeh, glyphs are visual markers or shapes used to represent data points in a plot. They are the fundamental building blocks for creating visualizations. Bokeh provides a variety of glyphs such as circles, squares, triangles, lines, and more.


```python
p = figure()


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.circle(x, y, size=10, color="blue")


p.line(x, y, line_width=2, color="red")


show(p)
```

## **Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?**

```python
p = figure(title = 'data')
p.xaxis.axis_label = "x-axis"
p.yaxis.axis_label = "y-axis"
p.scatter(x,y, color = 'red', legend_label = 'red points', size = 15)
show(p)
```

### A Bokeh server is a component of the Bokeh library that allows you to create and deploy interactive data applications and dashboards. It enables you to build dynamic and interactive plots that can be updated in real time based on user interactions, streaming data, or periodic updates.
### When using the Bokeh server, your Bokeh plots and visualizations are served as web applications. The server handles communication between the web browser and the Python runtime, allowing for bidirectional interaction. This means that users can interact with the plots in the browser, and the server can respond by updating the plots dynamically.

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
```

```python

output_file("line.html")

p = figure(title = 'data')
p.xaxis.axis_label = "x-axis"
p.yaxis.axis_label = "y-axis"
p.scatter(x,y, color = 'red', legend_label = 'red points', size = 15)
show(p)
```
